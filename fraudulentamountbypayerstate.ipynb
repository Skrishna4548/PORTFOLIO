```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv(r"C:\Users\Shyam Krishna. M\Downloads\Fraud Analytics Dataset1.csv")
```

```python
#group the fraudulent data by 'payer_state' and sum up 'payee_settlement_amount'
fraudulent_amount_by_payer_state = data.groupby('payer_state')['payee_settlement_amount'].sum().reset_index()
#sort the resulting DataFrame by 'payee_settlement_amount'in descending order
fraudulent_amount_by_payer_state = fraudulent_amount_by_payer_state.sort_values(
    by='payee_settlement_amount',ascending=False)
#initialize a figure with specified sizze for the plot 
plt.figure(figsize=(12,6))
#create a bar plot showing total fraudelent transactin amount by payer state
sns.barplot(x='payer_state',y='payee_settlement_amount',
            data=fraudulent_amount_by_payer_state)
#set the title of theplot and labels for the x and y axes 
plt.title('Fraudulent Transaction Amount by the Payer State')
plt.xlabel('payer state')
plt.ylabel('Total Fraudulent transaction Amount')
#rotate the x axis labels (states) for better readability
plt.xticks(rotation=90)
plt.show()
```
