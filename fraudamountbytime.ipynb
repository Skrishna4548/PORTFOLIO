```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

```

```python
data=pd.read_csv(r"C:\Users\Shyam Krishna. M\Downloads\Fraud Analytics Dataset1.csv")
```

```python
#Aggregate fraudulent amount by time of day 
fraudulent_amount_by_time = data.groupby('time_of_day')[
'payee_settlement_amount'].sum().reset_index()
#initislize pie chart figure with specified size 
plt.figure(figsize=(8,8))

#create pie chart for fraudulent transaction amounts by  time of day
plt.pie(fraudulent_amount_by_time['payee_settlement_amount'],
        labels=fraudulent_amount_by_time['time_of_day'],
        autopct='%1.1f%%',startangle=140)
#title for pie chart

plt.title('Distribution of Fraudulent Transaction Amounts byTime of Day')
plt.show()
```
