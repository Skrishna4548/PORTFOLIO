```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns 

```

```python
data = pd.read_csv(r"C:\Users\Shyam Krishna. M\Downloads\Fraud Analytics Dataset1.csv")
```

```python
#sum fraudulent amounts by credit type 
fraudulent_amount_by_credit_type= data.groupby('cred_type')['payee_settlement_amount'].sum().reset_index()
#initialize plot with specified fiq size
plt.figure(figsize=(10,6))
#create bar plot for fraudulent amounts bu credit type 
sns.barplot(x='cred_type',y='payee_settlement_amount',
            data=fraudulent_amount_by_credit_type)
#set ploot title and labels for clarity
plt.title('Total Amount of Fraudulent Transaction by Credit Type')
plt.xlabel('Credit Type')
plt.ylabel('Total Amount')
plt.show ()
```
